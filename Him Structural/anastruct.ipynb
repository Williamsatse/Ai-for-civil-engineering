```python
from anastruct import SystemElements
```

```python
ss = SystemElements(EA=15000, EI=5000)
```

```python
ss.add_element(location=[[0, 0], [5, 0]])
```

```python
ss.add_support_hinged(node_id=1)  # Articulé à gauche
ss.add_support_roll(node_id=2, direction=2)  # Mobile à droite (bloque y)
```

```python
ss.q_load(element_id=1, q=-10)  # Charge uniforme -10 kN/m
#ss.point_load(node_id=2, Fy=-30)  # Charge ponctuelle -30 kN à droite
```

```python
ss.solve()
```

```python
ss.show_structure()  # Vue globale
```

```python
ss.show_reaction_force()
```

```python
ss.show_shear_force()
```
