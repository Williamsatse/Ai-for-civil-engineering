```python
model = FEModel3D()
```

```python
# 2. Matériau
E = 200000  # MPa = kN/m²
G = 76923
nu = 0.3
model.add_material('Acier', E, G, nu)
```

```python
# 3. Section (rectangulaire 20×40 cm)
A = 0.08
Iy = (0.2 * 0.4**3) / 12   # ≈ 0.0010667 m⁴ → fort axe pour Mz
Iz = (0.4 * 0.2**3) / 12   # ≈ 0.0002667 m⁴
J = Iy + Iz                # approx
model.add_section('Rect', A, Iy, Iz, J)
```

```python
# 4. Nœuds
model.add_node('N1', 0.0, 0, 0)
model.add_node('N2', 5.0, 0, 0)
```

```python
model.add_member('M1', 'N1', 'N2', 'Acier', 'Rect')
```

```python
# 7. Ajouter les charges
# Charge répartie uniforme descendante sur tout le membre (Fy négatif = vers le bas)
beam.add_member_dist_load('M1', 'Fy', w1=-10, w2=-10)  # -10 kN/m sur toute la longueur
```

```python
# Charge ponctuelle à l'extrémité droite (N2)
beam.add_node_load('N2', 'FY', -30)  # -30 kN vers le bas
```
